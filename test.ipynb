### Matplotlib을 이용한 데이터 시각화 예제

다음 코드는 `matplotlib.pyplot`을 사용하여 간단한 선 그래프를 그리는 방법을 보여줍니다. `numpy`를 사용하여 예시 데이터를 생성합니다.

```python
# Install Korean font (NanumGothic) for Colab
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf

import matplotlib.pyplot as plt
import numpy as np

# Configure Matplotlib to use the Korean font
plt.rc('font', family='NanumGothic') # For Linux/Colab
plt.rcParams['axes.unicode_minus'] = False # To prevent minus sign from breaking

# 예시 데이터 생성
x = np.linspace(0, 10, 100) # 0부터 10까지 100개의 점
y = np.sin(x) # x 값에 대한 사인 함수

# 그래프 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.plot(x, y, label='sin(x) 함수') # 선 그래프 그리기
plt.title('간단한 Matplotlib 선 그래프') # 그래프 제목
plt.xlabel('X 축') # X축 레이블
plt.ylabel('Y 축') # Y축 레이블
plt.legend() # 범례 표시
plt.grid(True) # 그리드 표시
plt.show() # 그래프 보여주기
```
